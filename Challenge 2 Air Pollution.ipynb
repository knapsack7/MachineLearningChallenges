```python
import sklearn
import numpy as np
import pandas as pd
```

```python
X_train=pd.read_csv("Train.csv")
print(X_train)
print(X_train.shape)
```

```python
Y_test=pd.read_csv("Test.csv")
print(Y_test.columns)
print(Y_test.shape)
testData=Y_test
#print(Y_test)
sampleSubmission=pd.read_csv("sample_file.txt")
print(sampleSubmission.columns)
```

```python
X=X_train.values
data=X[:,:5]
#print(X)
#print(data.shape)
#print(data)
target=X[:,-1]
#print(target.shape)
#print(target)
```

```python
from sklearn.linear_model import LinearRegression #To create Linear Regression Object
from sklearn.model_selection import train_test_split #to split the data in train and test data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(data,target,test_size=0.1)# 20 percent of data we want in training set
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# 1.Create Linear Regression Object
lr=LinearRegression(normalize=True)
# 2.Training
lr.fit(X_train,Y_train)
# 3.output parameters (thetas)
print(lr.coef_) # coefficients from theta1 to theta13  
print(lr.intercept_)# theta0 (ie.intercept)
```

```python
print("Training score is %.4f"%(lr.score(X_train,Y_train)))
print("Testing score is %.4f"%(lr.score(X_test,Y_test)))
```

```python
output=lr.predict(testData)
print(output.shape)
#print(output)
```

```python
df=pd.DataFrame(data=output,columns=['Hello World'])
df.to_csv("PredictionAirPollution.csv",index=True)
```
